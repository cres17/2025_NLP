```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
```

```python
import openai
from openai import OpenAI

client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
upload_response = client.files.create(
    file=open("king-style-chat.jsonl", "rb"),
    purpose = 'fine-tune'
)

print(f"Upload Response : {upload_response}")

list_response = client.files.list()
print(f"List Response : {list_response}")

file_id = upload_response.id
retrieve_response  = client.files.retrieve(file_id)
print(f"Retrieve Response : {retrieve_response}")

delete_response = client.files.delete(file_id)
print(f"Delete Response : {delete_response}")
```
