```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
```

```python
from openai import OpenAI

client = OpenAI(api_key= OPENAI_API_KEY)
```

```python
myfile = client.files.create(
    file = open("heartspring-style-chat-light.jsonl", "rb"),
    purpose = "fine-tune"
)

print(myfile.id)
```

```python
myjob = client.fine_tuning.jobs.create(
    training_file = myfile.id,
    model = "gpt-3.5-turbo"
)

print(myjob.id)
```

```python
print(myjob.status)
```

```python
light_fine_tuned_model_id = "ft:gpt-3.5-turbo-0125:personal::CjLXDQIM"
model_id = "gpt-3.5-turbo-0125"
```

```python
test_questions = [
    "오늘 날씨 너무 춥지 않아요?",
    "주말에는 뭐하세요?",
    "요즘 관심사는 뭐에요?",
    "물가 너무 많이 오르지 않았어?",
    "이번 달 월급 다 썼는데 사고 싶은 건 아직 많아.. 어떡하지?",
    "행복은 뭐라고 생각해?",
    "집에 가고 싶다",
    "클라이밍 재밌어보이지 않아?",
    "무슨 영화 좋아해?",
    "커피는 너무 쓰지 않아?"
]
```

```python
# 일상 대화 테스트 실행
for i, question in enumerate(test_questions):
    print(f"--- 질문 {i+1} ---")
    print(f"User: {question}")

    completion = client.chat.completions.create(
        model=light_fine_tuned_model_id,
        messages=[
            {"role": "system", "content": "하츄핑의 말투를 사용하여 답장하시오."},
            {"role": "user", "content": question}
        ]
    )

    # 답변 출력
    answer = completion.choices[0].message.content
    print(f"Hachuping: {answer}\n")
```
